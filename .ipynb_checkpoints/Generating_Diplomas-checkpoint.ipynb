```python
# Read all the information
import pandas as pd
df = pd.read_csv("./Curso-de-Python-FutureLabs.csv", encoding="latin1")
df
```

```python
# Calculating the necessary points to 'deserve' a diploma.
necessary_points = 500 + 200 + 600 + 100 + 500 + 100 + 200
necessary_points *= 0.4
necessary_points
```

```python
# Listing all the participants that have at least that number of points.
names = []
for i in df.index:
    suma = sum(df.iloc[i][2:])
    if suma >= necessary_points and not pd.isnull(df.iloc[i][1]):
        names.append(df.iloc[i][1])
    #print(i)
print(names)
print(len(names))
```

```python
# Creating a new data_frame but is not necessary.
columns = ["names"]
data_names = pd.DataFrame(names, columns=columns)
# Cleaning data:
data_names["names"][3] = "Andrea Estefania Jimenez Saldaña"
data_names["names"][42] = data_names["names"][42][:-3]
data_names
```

```python
import os
import math
# Definition of things
size = 450
x = (diploma.size[0])/2
y = 1250
font_path = os.environ.get("font_path") + "FredokaOne-Regular.ttf"


for name in data_names["names"]:
    # Definition of things
    name = " ".join([n.capitalize() for n in name.split(" ")])
    numLetters = len(name)
    
    # Resizing
    new_size = size * 0.97**numLetters
    new_size = int(new_size)
    
    if numLetters <= 5:
        new_size = 300
    
    # Positioning
    pos_x = x - numLetters*new_size/4

    pos_y = y
    if numLetters >= 18:
        pos_y = y + new_size/3 
    if numLetters >= 20:
        pos_x = x - numLetters*new_size/4 - 100
    
    print(name, pos_x, pos_y, numLetters, new_size, numLetters*new_size/2)
        

    
    # Diploma writing.
    diploma01 = diploma.copy()
    draw = ImageDraw.Draw(diploma01)
    font = ImageFont.truetype(font_path, new_size)
    draw.text((pos_x,pos_y), name, (255, 206, 76), font=font)
    diploma01.save(f"./Diplomas/{name}.jpg")
#   diploma01.show()
#    plt.imshow(diploma01)
```
